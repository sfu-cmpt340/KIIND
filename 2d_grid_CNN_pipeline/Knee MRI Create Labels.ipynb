```python
import csv
import os
import numpy as np

# there will be overlap as data can be all of acl/meniscus/abnormal, 
#therefore will create total of 7 classes to encapsulate all cases
#this code creates a single labels csv that combines information from all acl,meniscus and abnormal label csvs
#make sure to update file path for csvs and for file to be saved




dataArray = [[], #number
            [], #acl
            [], #meniscus
            []] #abnormal
count = 0

# acl
with open('MRNet-v1.0/valid-acl.csv', newline='') as csvfile:
    csv_reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in csv_reader:
        if len(row) != 0:
            dataArray[0].append(row[0]) 
            dataArray[1].append(int(row[1])) 
            count += 1

# meniscus
count = 0
with open('MRNet-v1.0/valid-meniscus.csv', newline='') as csvfile:
    csv_reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in csv_reader:
        if len(row) != 0:
            dataArray[2].append(int(row[1])) 
            count += 1

# abnormal
count = 0
with open('MRNet-v1.0/valid-abnormal.csv', newline='') as csvfile:
    csv_reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in csv_reader:
        if len(row) != 0:
            dataArray[3].append(int(row[1]))
            count += 1

# print(dataArray[1][1])
# print(str(dataArray[1][1]).zfill(4))


#at end labels will either 0,1,2,3,4,5,6,7
    #0 = healthy
    #1 = acl tear ONLY
    #2 = meniscus tear ONLY
    #3 = abnormal ONLY
    #4 = acl & meniscus 
    #5 = meniscus & abnormal
    #6 = acl & abnormal
    #7 = acl & meniscus & abnormal

nSamples = len(dataArray[0])
labels = np.empty(nSamples, dtype=object)

for i in range(nSamples):
    ################2 classes: healthy (0) or not healthy (1), not healthy can be acl,meniscus or abnormal)
     if (dataArray[1][i] == 0 and dataArray[2][i] == 0 and dataArray[3][i] == 0): #volume is healthy ONLY
        labels[i] = 0
     else:
        labels[i] = 1
        
    
    
   

###################7 classes###############    
#     #single pathology
#     if (dataArray[1][i] == 0 and dataArray[2][i] == 0 and dataArray[3][i] == 0): #volume is healthy ONLY
#         labels[i] = 0
#     if (dataArray[1][i] == 1 and dataArray[2][i] == 0 and dataArray[3][i] == 0): #volume has acl tear ONLY
#         labels[i] = 1
#     if (dataArray[1][i] == 0 and dataArray[2][i] == 1 and dataArray[3][i] == 0): #volume has meniscus tear ONLY
#         labels[i] = 2    
#     if (dataArray[1][i] == 0 and dataArray[2][i] == 0 and dataArray[3][i] == 1): #volume is abnormal ONLY
#         labels[i] = 3
        
#     #combinations of pathologies
#     if (dataArray[1][i] == 1 and dataArray[2][i] == 1 and dataArray[3][i] == 0): #volume has acl tear AND meniscus tear
#         labels[i] = 4
#     if (dataArray[1][i] == 0 and dataArray[2][i] == 1 and dataArray[3][i] == 1): #volume has meniscus tear AND abnormal 
#         labels[i] = 5
#     if (dataArray[1][i] == 1 and dataArray[2][i] == 0 and dataArray[3][i] == 1): #volume has acl tear AND abnormal
#         labels[i] = 6
#     if (dataArray[1][i] == 1 and dataArray[2][i] == 1 and dataArray[3][i] == 1): #volume has acl AND meniscus AND abnormal
#         labels[i] = 7
     

    

#save labels as a csv
labels = [int(x) for x in labels]
csv_name = "valid-MRI-labels_healthy_non_healthy.csv"
np.savetxt(csv_name, labels, delimiter=",", fmt = '%d')
print("Saved csv:", csv_name, "in path", os.getcwd())
```
